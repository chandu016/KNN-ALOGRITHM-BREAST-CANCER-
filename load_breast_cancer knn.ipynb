## Import the required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cancer = load_breast_cancer()
```

## CONVERT RAW DATA(ARRAY) TO DATA FRAME

```python
df_feat = pd.DataFrame(cancer['data'], columns =cancer['feature_names'])
```

## CONVERT THE DATA TO STANDARD FORM  Z = (X-µ)/σ

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df_feat)
```

```python
scaled_feature = scaler.transform(df_feat)
```

```python
df_feat_scaled = pd.DataFrame(scaled_feature,columns = df_feat.columns )
```

```python

    
    df_feat_scaled
```

## SPLIT THE DATA INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_feat_scaled, np.ravel(df_target), test_size = 0.3, random_state =0)
```

## IMPORT THE KNN LIBRARY

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors =1)
```

```python
knn.fit(x_train,y_train)
```

```python
pred =knn.predict(x_test)
```

```python
pred
```

## CHECK THE ACCURACY OF MODEL

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

# how to choose k value

```python
training_accuracy =[]
testing_accuracy =[]

neighbors = range(1,40)

for no_of_neighbors in neighbors:
    kn =KNeighborsClassifier(n_neighbors=no_of_neighbors)
    kn.fit(x_train,y_train)
    pred_i = kn.predict(x_test)
    training_accuracy.append(np.mean(pred_i != y_test))
    testing_accuracy.append(np.mean(pred_i != y_test))

```

## TO CHECK THE LESS ERROR RATE FOR N_NEIGHBOUR

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),training_accuracy,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),testing_accuracy,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
